```python
import pandas as pd
```

```python

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print("--- İLK 5 SATIR ---")
print(df.head())
```

```python
print("\n--- VERİ SETİ BİLGİSİ ---")
print(df.info())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print("Sütunun Yeni Veri Tipi:", df['TotalCharges'].dtype)
```

```python
print("Oluşan Eksik Değer Sayısı:", df['TotalCharges'].isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
print("Kalan Toplam Eksik Değer Sayısı:", df.isnull().sum().sum())
```

```python
print("Veri Setinin Yeni Boyutu (Satır, Sütun):", df.shape)
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
```

```python
print(df[['gender', 'Churn']].head())
```

```python
X = df[['gender', 'MonthlyCharges', 'TotalCharges']]
```

```python
y = df['Churn']
```

```python
print("X (Özellikler) Boyutu:", X.shape)
```

```python
print("y (Hedef) Boyutu:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print("X_train (Eğitim Özellikleri):", X_train.shape)
print("X_test (Test Özellikleri):", X_test.shape)
print("y_train (Eğitim Hedefi):", y_train.shape)
print("y_test (Test Hedefi):", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print("İlk yapay zeka modelin başarıyla eğitildi Eren! Şimdi test zamanı.")
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Eren'in İlk Modelinin Doğruluk (Accuracy) Skoru: %{accuracy * 100:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print("--- KARMAŞIKLIK MATRİSİ ---")
print(cm)
```

```python
print("\n--- DETAYLI SINIFLANDIRMA RAPORU ---")
print(classification_report(y_test, y_pred))
```

```python
df_clean = df.drop(columns=['customerID'])
```

```python
df_encoded = pd.get_dummies(df_clean, drop_first=True)
```

```python
print("Yeni Genişletilmiş Veri Setinin Boyutu (Satır, Sütun):")
print(df_encoded.shape)
```

```python
X_yeni = df_encoded.drop(columns=['Churn'])
y_yeni = df_encoded['Churn']
```

```python
X_train_yeni, X_test_yeni, y_train_yeni, y_test_yeni = train_test_split(
    X_yeni, y_yeni, test_size=0.20, random_state=42)
```

```python
model_yeni = LogisticRegression(max_iter=1500)
model_yeni.fit(X_train_yeni, y_train_yeni)
```

```python
y_pred_yeni = model_yeni.predict(X_test_yeni)
```

```python
print("--- YENİ DETAYLI SINIFLANDIRMA RAPORU ---")
print(classification_report(y_test_yeni, y_pred_yeni))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train_yeni)
X_test_scaled = scaler.transform(X_test_yeni)
```

```python
model_final = LogisticRegression(max_iter=1000)
model_final.fit(X_train_scaled, y_train_yeni)
```

```python
y_pred_final = model_final.predict(X_test_scaled)
```

```python
print("--- ÖLÇEKLENMİŞ VE GÜÇLENDİRİLMİŞ YENİ RAPOR ---")
print(classification_report(y_test_yeni, y_pred_final))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf_balanced.fit(X_train_scaled, y_train_yeni)
```

```python
y_pred_balanced = rf_balanced.predict(X_test_scaled)
```

```python
print("--- ⚖️ DENGELENMİŞ RANDOM FOREST RAPORU ⚖️ ---")
print(classification_report(y_test_yeni, y_pred_balanced))
```
